> <h3><strong>Método de tuckey</strong></h3>

```python
def tuckey_method(X):
  
    X = np.array(X)

    Q1 = np.quantile(X,0.25)
    Q3 = np.quantile(X,0.75)

    IQR = Q3-Q1
    limIn = Q1-(IQR*1.5)
    limSp = Q3+(IQR*1.5)
    
    return limIn, limSp, np.where((X>limSp)|(X<limIn))
```
